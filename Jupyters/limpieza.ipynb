```python
# Diccionario meses español → número
meses = {
    'enero': '01', 'febrero': '02', 'marzo': '03', 'abril': '04',
    'mayo': '05', 'junio': '06', 'julio': '07', 'agosto': '08',
    'septiembre': '09', 'setiembre': '09', 'octubre': '10',
    'noviembre': '11', 'diciembre': '12'
}

# Limpiar y normalizar texto
df['date'] = df['date'].astype(str).str.strip().str.lower()

# Quitar tildes
tildes = {'á': 'a', 'é': 'e', 'í': 'i', 'ó': 'o', 'ú': 'u'}
for acento, sin_acento in tildes.items():
    df['date'] = df['date'].str.replace(acento, sin_acento, regex=False)

# Reemplazar nombre del mes por número
for mes, num in meses.items():
    df['date'] = df['date'].str.replace(mes, num, regex=False)

# Crear nueva columna como fecha
df['date_dt'] = pd.to_datetime(df['date'], format='%d-%m-%Y', errors='coerce')

# Formatear la fecha como dd-mm-aaaa en esa columna
df['date_dt'] = df['date_dt'].dt.strftime('%d-%m-%Y')
```

```python
df.drop('date', axis = 1, inplace=True )
```

```python
sp.convertir_columnas(df)
```

```python
df['year'] = df['date_dt'].dt.year
df['month'] = df['date_dt'].dt.month
df['day'] = df['date_dt'].dt.day_name() 
df['quarter'] = df['date_dt'].dt.quarter
```
